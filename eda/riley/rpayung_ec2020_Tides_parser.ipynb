```python
import pandas as pd;
import numpy as np;
```

```python
tidesdata = pd.read_csv("TIDES_newdata.csv");
ec2020 = pd.read_csv("EC2020.csv");
```

```python
LastUID = int(str(ec2020['UniqueID'][-1:]).split('-')[1][0:5])
LastRSID = int(ec2020['RecordSequenceID'][-1:]);
LastOID = int(ec2020['OBJECTID'][-1:])

def month(dateMM):
    dateMM=int(dateMM);
    if (dateMM == 1):
        return "Jan";
    elif(dateMM == 2):
        return "Feb";
    elif(dateMM == 3):
        return "Mar";
    elif(dateMM == 4):
        return "Apr";
    elif(dateMM == 5):
        return "May";
    elif(dateMM == 6):
        return "Jun";
    elif(dateMM == 7):
        return "Jul";
    elif(dateMM == 8):
        return "Aug";
    elif(dateMM == 9):
        return "Sep";
    elif(dateMM == 10):
        return "Oct";
    elif(dateMM == 11):
        return "Nov";
    else:
        return "Dec";
    
for i in tidesdata:
    lon = i['GPS'].split(",")[0];
    lat = i['GPS'].split(",")[1];
    try:
        countryCode = i['State'].split(',')[1];
    except:
        countryCode = "";
    try:
        state = i['State'].split(",")[0]
    except:
        state = "";
    try:
        county = i['Zone'].split(",")[0]
    except:
        county = "";
    dateDD = i['Cleanup Date'].split('/')[1];
    dateMM = i['Cleanup Date'].split('/')[0];
    dateYY = i['Cleanup Date'].split('/')[2]
    newData = {
        'X':lon,
        'Y':lat,
        'ObjectID':LastOID+1,
        'RecordSequenceID':LastRSID+1,
        'UniqueID':'TID-'+str(LastUID+1),
        'SourceID':i['Cleanup ID'],
        'LocationFreqID':'TIDES (Lon'+lon+'Lat '+lat+")",
        'Location':i['Zone'],
        'Dataset':'Ocean Conservancy TIDES Report',
        'Orginization':i['Group Name'],
        'Other':"",
        'CountryName_FromSource':countryCode,
        'SubCountry_L1_FromSource':state,
        'SubCountry_L2_FromSource':county,
        'Longitude1':lon,
        'Latitude1':lat,
        'Longitude2':"",
        'Latitude2':"",
        'TotalWidth_m':"",
        'TotalLength_m':float(i['Miles'])/0.00062137,
        'TotalArea_Sq_m':"",
        'ShorelineName':"",
        'WaterfrontName':"",
        'BeachAreaLandcover':"",
        'BeachType':"",
        'EventType':i['Cleanup Type'],
        'TotalVolunteers':i['People'],
        'DateOriginal':str(dateMM + '/' + dateDD+'/'+dateYY+' 0:00'),
        'DateStandardized':str(dateMM + '/' + dateDD+'/'+dateYY+' 0:00'),
        'MonthYear':str(dateMM+'/'+dateYY),
        'Year':str(dateYY),
        'MonthNum':str(dateMM),
        'Month':str(month(dateMM)),
        'Day':str(dateDD),
        'StartTime':"",
        'DOW':"",
        'FieldObservations':"",
        'DebrisDescription':"",
        'TotalItems_EventRecord':i['Total Items Collected'],
        'TotalClassifiedItems_EC2020':i['Total Items Collected'],
        'PCT_PlasticAndFoam':(i['Plastic Pieces']+i['Foam Pieces'])/i['Total Items Collected']*100,
        'PCT_Glass_Rubber_Lumber_Metal':(i['Construction Materials']+i['Glass Pieces'])/i['Total Items Collected']*100,
        'SUM_Hard_PlasticsBeverageBottle': i['Beverage Bottles (Plastic)'],
        'SUM_Hard_OtherPlasticBottle': i['Other Plastic Bottles (oil, bleach, etc.)'],
        'SUM_HardOrSoft_PlasticBottleCap': i['Bottle Caps (Plastic)'],
        'SUM_PlasticOrFoamFoodContainer': i['Take Out/Away Containers (Foam)'],
        'SUM_Hard_BucketOrCrate':0,
        'SUM_Hard_Lighter':i['Cigarette Lighters'],
        'SUM_OtherHardPlastic':i['Strapping Bands']+i['Cups, Plates (Plastic)'],
        'SUM_PlasticOrFoamPlatesBowlsCup':i['Cups, Plates (Plastic)']+i['Cups, Plates (Foam)'],
        'SUM_HardSoft_PersonalCareProduc':i['Condoms']+i['Syringes']+i['Tampons/Tampon Applicators']+i['Gloves & Masks (PPE)']+i['Personal Hygiene (Clean Swell)'] + i['Diapers'],
        'SUM_HardSoftLollipopStick_EarBu':0,
        'SUM_Soft_Bag':i['Grocery Bags (Plastic)']+i['Other Plastic Bags']+i['# of bags'],
        'SUM_Soft_WrapperOrLabel':i['Food Wrappers (candy, chips, etc.)'],
        'SUM_Soft_Straw':i['Straws, Stirrers'],
        'SUM_Soft_OtherPlastic':i['6-Pack Holders']+i['Other Plastic/Foam Packaging']+i['Other Packaging (Clean Swell)'],
        'SUM_Soft_CigaretteButts':i['Cigarette Butts'],
        'SUM_Soft_StringRingRibbon':0,
        'Fishing_Net':i['Fishing Net & Pieces'],
        'SUM_FishingLineLureRope':i['Fishing Line (1 yard/meter = 1 piece)'] + i['Rope (1 yard/meter = 1 piece)'],
        'Fishing_BuoysAndFloats':i['Fishing Buoys, Pots & Traps'],
        'SUM_Foam_OtherPlasticDebris':i['Other Plastic/Foam Packaging']+i['Tobacco Packaging/Wrap']+i['Beverage Sachets']+i['Cigar Tips'],
        'SUM_OtherPlasticDebris':0,
        'NAME':state,
        'COUNTRY':i['Country'],
        'ISO_CODE':str(countryCode[:2]+state[:1]+state[-1:]),
        'ISO_CC':str(countryCode[:2]),
        'ISO_SUB':str(state[:1]+state[-1:]),
        'ADMINTYPE':'State',
        'DISPUTED':0,
        'NOTES':"",
        'AUTONOMOUS':0,
        'COUNTRYAFF':i['Coutry'],
        'CONTINENT':"",
        'LAND_TYPE':'',
        'LAND_RANK':5,
        'Shape__Area':0,
        'Shape__Length':0,
        'Count_':1,
        'Soft_Sheets2':0,
        'PlasticStraps2':i['Strapping Bands'],
        'FishingGlowSticks2':i['Fishing Gear (Clean Swell)'],
        'FishingOtherPlasticDebris2':0
    };
    LastRSID += 1;
    LastUID += 1;
    LastOID += 1;
```
