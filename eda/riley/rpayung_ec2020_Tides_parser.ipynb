```python
import pandas as pd;
import numpy as np;
```

```python
def month(dateMM):
    dateMM=int(dateMM);
    if (dateMM == 1):
        return "Jan";
    elif(dateMM == 2):
        return "Feb";
    elif(dateMM == 3):
        return "Mar";
    elif(dateMM == 4):
        return "Apr";
    elif(dateMM == 5):
        return "May";
    elif(dateMM == 6):
        return "Jun";
    elif(dateMM == 7):
        return "Jul";
    elif(dateMM == 8):
        return "Aug";
    elif(dateMM == 9):
        return "Sep";
    elif(dateMM == 10):
        return "Oct";
    elif(dateMM == 11):
        return "Nov";
    else:
        return "Dec";


def tidesparse(ecCSV = 'EC2020.csv',TidesNew = 'TIDES_newdata.csv',output='EarthChallengeNew.csv'):
    
    tidesdata = pd.DataFrame(pd.read_csv(TidesNew));
    ec2020 = pd.DataFrame(pd.read_csv(ecCSV));
    LastUID = int(str(ec2020['UniqueID'][-1:]).split('-')[1][0:5])
    LastRSID = int(ec2020['RecordSequenceID'][-1:]);
    LastOID = int(ec2020['OBJECTID'][-1:])

    a = 0;

    for i in range(len(tidesdata)):
        if (tidesdata['Cleanup Date'][i] == ['nan']):
            break;
        GPS = tidesdata['GPS'][i]
        lat = 0;
        lon = 0;
        try:
            GPS = str(GPS).split(',')
            lat = float(GPS[0]);
            lon = float(GPS[1]);
        except:
            lat = 0;
            lon = 0;
        try:
            countryCode = i['State'].split(',')[1];
        except:
            countryCode = "";
        try:
            state = i['State'].split(",")[0]
        except:
            state = "";
        try:
            county = i['Zone'].split(",")[0]
        except:
            county = "";
        try:
            dateDD = str(tidesdata['Cleanup Date'][i]).split('/')[1];
            dateMM = str(tidesdata['Cleanup Date'][i]).split('/')[0];
            dateYY = str(tidesdata['Cleanup Date'][i]).split('/')[2]
        except:
            break;
        newData = {
            'X':lon,
            'Y':lat,
            'ObjectID':LastOID+1,
            'RecordSequenceID':LastRSID+1,
            'UniqueID':'TID-'+str(LastUID+1),
            'SourceID':tidesdata['Cleanup ID'][i],
            'LocationFreqID':'TIDES (Lon'+str(lon)+'Lat '+str(lat)+")",
            'Location':tidesdata['Zone'][i],
            'Dataset':'Ocean Conservancy TIDES Report',
            'Orginization':tidesdata['Group Name'][i],
            'Other':"",
            'CountryName_FromSource':countryCode,
            'SubCountry_L1_FromSource':state,
            'SubCountry_L2_FromSource':county,
            'Longitude1':lon,
            'Latitude1':lat,
            'Longitude2':"",
            'Latitude2':"",
            'TotalWidth_m':"",
            'TotalLength_m':float(tidesdata['Miles'][i])/0.00062137,
            'TotalArea_Sq_m':"",
            'ShorelineName':"",
            'WaterfrontName':"",
            'BeachAreaLandcover':"",
            'BeachType':"",
            'EventType':tidesdata['Cleanup Type'][i],
            'TotalVolunteers':tidesdata['People'][i],
            'DateOriginal':str(dateMM + '/' + dateDD+'/'+dateYY+' 0:00'),
            'DateStandardized':str(dateMM + '/' + dateDD+'/'+dateYY+' 0:00'),
            'MonthYear':str(dateMM+'/'+dateYY),
            'Year':str(dateYY),
            'MonthNum':str(dateMM),
            'Month':str(month(dateMM)),
            'Day':str(dateDD),
            'StartTime':"",
            'DOW':"",
            'FieldObservations':"",
            'DebrisDescription':"",
            'TotalItems_EventRecord':tidesdata['Total Items Collected'][i],
            'TotalClassifiedItems_EC2020':tidesdata['Total Items Collected'][i],
            'PCT_PlasticAndFoam':(tidesdata['Plastic Pieces'][i]+tidesdata['Foam Pieces'][i])/tidesdata['Total Items Collected'][i]*100,
            'PCT_Glass_Rubber_Lumber_Metal':(tidesdata['Construction Materials'][i]+tidesdata['Glass Pieces'][i])/tidesdata['Total Items Collected'][i]*100,
            'SUM_Hard_PlasticsBeverageBottle': tidesdata['Beverage Bottles (Plastic)'][i],
            'SUM_Hard_OtherPlasticBottle': tidesdata['Other Plastic Bottles (oil, bleach, etc.)'][i],
            'SUM_HardOrSoft_PlasticBottleCap': tidesdata['Bottle Caps (Plastic)'][i],
            'SUM_PlasticOrFoamFoodContainer': tidesdata['Take Out/Away Containers (Foam)'][i],
            'SUM_Hard_BucketOrCrate':0,
            'SUM_Hard_Lighter':tidesdata['Cigarette Lighters'][i],
            'SUM_OtherHardPlastic':tidesdata['Strapping Bands'][i]+tidesdata['Cups, Plates (Plastic)'][i],
            'SUM_PlasticOrFoamPlatesBowlsCup':tidesdata['Cups, Plates (Plastic)'][i]+tidesdata['Cups, Plates (Foam)'][i],
            'SUM_HardSoft_PersonalCareProduc':tidesdata['Condoms'][i]+tidesdata['Syringes'][i]+tidesdata['Tampons/Tampon Applicators'][i]+tidesdata['Gloves & Masks (PPE)'][i]+tidesdata['Personal Hygiene (Clean Swell)'][i] + tidesdata['Diapers'][i],
            'SUM_HardSoftLollipopStick_EarBu':0,
            'SUM_Soft_Bag':tidesdata['Grocery Bags (Plastic)'][i]+tidesdata['Other Plastic Bags'][i]+tidesdata['# of bags'][i],
            'SUM_Soft_WrapperOrLabel':tidesdata['Food Wrappers (candy, chips, etc.)'][i],
            'SUM_Soft_Straw':tidesdata['Straws, Stirrers'][i],
            'SUM_Soft_OtherPlastic':tidesdata['6-Pack Holders'][i]+tidesdata['Other Plastic/Foam Packaging'][i]+tidesdata['Other Packaging (Clean Swell)'][i],
            'SUM_Soft_CigaretteButts':tidesdata['Cigarette Butts'][i],
            'SUM_Soft_StringRingRibbon':0,
            'Fishing_Net':tidesdata['Fishing Net & Pieces'][i],
            'SUM_FishingLineLureRope':tidesdata['Fishing Line (1 yard/meter = 1 piece)'][i] + tidesdata['Rope (1 yard/meter = 1 piece)'][i],
            'Fishing_BuoysAndFloats':tidesdata['Fishing Buoys, Pots & Traps'][i],
            'SUM_Foam_OtherPlasticDebris':tidesdata['Other Plastic/Foam Packaging'][i]+tidesdata['Tobacco Packaging/Wrap'][i]+tidesdata['Beverages Sachets'][i]+tidesdata['Cigar Tips'][i],
            'SUM_OtherPlasticDebris':0,
            'NAME':state,
            'COUNTRY':tidesdata['Country'][i],
            'ISO_CODE':str(countryCode[:2]+state[:1]+state[-1:]),
            'ISO_CC':str(countryCode[:2]),
            'ISO_SUB':str(state[:1]+state[-1:]),
            'ADMINTYPE':'State',
            'DISPUTED':0,
            'NOTES':"",
            'AUTONOMOUS':0,
            'COUNTRYAFF':tidesdata['Country'][i],
            'CONTINENT':"",
            'LAND_TYPE':'',
            'LAND_RANK':5,
            'Shape__Area':0,
            'Shape__Length':0,
            'Count_':1,
            'Soft_Sheets2':0,
            'PlasticStraps2':tidesdata['Strapping Bands'][i],
            'FishingGlowSticks2':tidesdata['Fishing Gear (Clean Swell)'][i],
            'FishingOtherPlasticDebris2':0
        };
        LastRSID += 1;
        LastUID += 1;
        LastOID += 1;

        ec2020.append(newData, ignore_index=True)
        
    ec2020new = ec2020
    ec2020new.to_csv(output)
```
